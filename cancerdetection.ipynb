```python
import numpy as np
import pandas as pd
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
data_frame = pd.read_csv('data.csv')
data_frame.head()
```

```python
data_frame
```

```python
data = data_frame.copy()
```

```python
data.shape
```

```python
data.info()
```

```python
#Data of diagnosis is converted into Binary Data
df_one = pd.get_dummies(data['diagnosis'])
```

```python
df_one.head()
```

```python
#Binary Data is Concatenated into Dataframe
df_two = pd.concat((df_one, data), axis=1)
```

```python
df_two.head()
```

```python
# Diagnosiscolumn is dropped
df_two = df_two.drop(['diagnosis'], axis=1)
```

```python
# We want M =0 and B =1 So we drop M column here
df_two = df_two.drop(['B'], axis=1)
```

```python
df_two.head()
```

```python
#False = No Cancer & True = Cancer
data2= df_two.rename(columns={'M':'diagnosis'})
```

```python
data2
```

```python
data2 = data2.drop(columns='Unnamed: 32', axis=1)
```

```python
data2 = data2.drop(columns='id', axis=1)
```

```python
#statistical measures about data
data2.describe()
```

```python
#checking the distrubution of target variable
data2['diagnosis'].value_counts()
```

```python
#True = Malignant
#False = Benign

data2.groupby('diagnosis').mean() 
```

```python
#Separating the features and target

X = data2.drop(columns='diagnosis', axis=1)
Y = data2['diagnosis']
```

```python
#Splitting data into training data and test data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
(X.shape, X_train.shape, X_test.shape)
```

```python
X_train
```

```python
#Model Training 

#Logistic Regression

model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Model Evaluation

#Accuracy Score

#Accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)

```

```python
training_data_accuracy
```

```python
#Accuracy on testing data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
testing_data_accuracy
```
